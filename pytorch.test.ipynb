```python
import torch
from torch.optim import AdamW
from torch.nn import Linear
from transformers import get_scheduler

from torch.optim.lr_scheduler import LinearLR

# 1. 创建一个张量
x = Linear(10, 10)

optimizer = AdamW(x.parameters(), lr=1e-5)

# lr_scheduler = get_scheduler(
#     name="linear",
#     optimizer=optimizer,
#     num_warmup_steps=0,
#     num_training_steps=10,
# )

lr_scheduler = LinearLR(optimizer, start_factor=1.0, end_factor=0.1, total_iters=3)


print(optimizer)


optimizer.step()
lr_scheduler.step()
print(optimizer)


optimizer.step()
lr_scheduler.step()
print(optimizer)

optimizer.step()
lr_scheduler.step()
print(optimizer)

optimizer.step()
lr_scheduler.step()
print(optimizer)
```
